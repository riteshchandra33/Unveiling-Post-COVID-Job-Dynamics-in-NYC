```python
import pandas as pd
import matplotlib as mat
```

```python
df=pd.read_csv("/Users/riteshchandra/Desktop/Jobs_NYC_Postings_Cleaned.csv")
df.head()
```

```python
#Getting to know the nature and different statistics of the data
df.describe()
```

```python
#dropping the duplicates in jobID coulumn maaking it distinct
df.drop_duplicates(subset=['Job ID'], keep='first', inplace=True)
```

```python
#after the deletion of all the duplicate data in the jobID we return to look into see the change in dataset
df.describe()
```

```python
#printing the columns in the dataset
print(df.columns)
```

```python
#here we see that there are columns which repeat the same information and some columns are empty. To remove the redundency of data we drop certain columns
columns_to_drop = ['Business Title', 'Additional Information', 'Work Location 1','Process Date','Job Description','To Apply','Preferred Skills','Minimum Qual Requirements','Recruitment Contact']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
#Dataset after removal of all the columns
df.head()
```

```python
print(df.columns)
```

```python
#for the feasibility of the residency requirement, we replace the data as required, not required and hybrid categories in the dataset
df['Residency Requirement'] = df['Residency Requirement'].replace({
    'New York City residency is required. ': 'Required',
    'New York City residency is not required': 'Not required'
})
```

```python
#printing the Residency column from the dataset
df['Residency Requirement']
```

```python
#Replacing the values of 'Posting date' with 'Posting Updated'
df['Posting Date'] = df['Posting Updated']
df.drop(columns=['Posting Updated'], inplace=True)
```

```python
#printing the Posting date column
df['Posting Date']
```

```python
#Extracting the year of job postings from the postings date column
df['Job Postings Year'] = df['Posting Date'].str[-2:]
df['Job Postings Year'] = df['Job Postings Year'].replace({
    '23': '2023',
    '22': '2022',
    '21':'2021'
})
```

```python
#Displaying the postings year
df['Job Postings Year']
```

```python
#generating a new .csv file which will be input to R for further visualizations and model creation
df.to_csv('Jobs_NYC_Postings_.csv', index=False)
```
